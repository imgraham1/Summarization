```
import statistics
from summa.summarizer import summarize
```

```
fh = open('y_test_full.txt')
gold = []
read_lines = fh.readlines()
for line in read_lines:
    gold.append(line)
fh.close()

fh = open('X_test_full.txt')
articles = []
read_lines = fh.readlines()
for line in read_lines:
    articles.append(line)
fh.close()
```

```
summaries = [summarize(story, words=60) for story in articles]
```

```
with open("extractive_summaries.txt", "w") as filehandle:
    for listitem in summaries:
        filehandle.write('%s\n' % listitem)
```
