```
fh = open('x_train_filtered.txt')
x_train = []
read_lines = fh.readlines()
for line in read_lines:
    x_train.append(line)
fh.close()

fh = open('X_test_full.txt')
x_test = []
read_lines = fh.readlines()
for line in read_lines:
    x_test.append(line)
fh.close()

fh = open('y_train_filtered.txt')
y_train = []
read_lines = fh.readlines()
for line in read_lines:
    y_train.append(line)
fh.close()

fh = open('y_test_full.txt')
y_test = []
read_lines = fh.readlines()
for line in read_lines:
    y_test.append(line)
fh.close()
```

```
src_vocab = []

for x in x_train:
  x = x.split()
  for i in x:
    src_vocab.append(i)


tgt_vocab = []
for x in y_train:
  x = x.split()
  for i in x:
    tgt_vocab.append(i)

print(len(src_vocab))

src_vocab = list(set(src_vocab))
tgt_vocab = list(set(tgt_vocab))

print(len(src_vocab))
```

```
print(len(src_vocab))
print(len(tgt_vocab))
```

```
with open("OpenNMT-py/pointer_gen/src_vocab_mine.txt", "w") as filehandle:
    for listitem in src_vocab:
        filehandle.write('%s\n' % listitem)

with open("OpenNMT-py/pointer_gen/tgt_vocab_mine.txt", "w") as filehandle:
    for listitem in tgt_vocab:
        filehandle.write('%s\n' % listitem)
```
